```python
df = pd.read_csv('global_superstore_2016.csv', encoding='latin1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#converting 'Order Date' and 'Ship Date' to datetime objects
order_date = df['Order Date'].tolist()
ship_date = df['Ship Date'].tolist()
order_datetime = pd.to_datetime(order_date)
ship_datetime = pd.to_datetime(ship_date)

print(order_datetime)

print(ship_datetime)
```

```python
Order_Date = pd.Series(order_datetime, name='Order_Date')
Ship_Date = pd. Series(ship_datetime, name='Ship_Date')
```

```python
df.columns.values
```

```python
# Drop unwanted columns
df.drop(columns=['Row ID','Order Date','Ship Date','Customer ID','Postal Code'], inplace=True)
```

```python
df.head()
```

```python
data = pd.concat([df,Order_Date,Ship_Date], axis=1)
data.head()
```

```python
#checking columns 'Sales' and 'Profit' for $ symbol
if data['Sales'].str.contains('$').any():
  print("yes, $ is present in column 'Sales'")

if data['Profit'].str.contains('$').any():
  print("yes, $ is present in column 'Profit'") 
```

```python
#Cleaning columns 'Sales' and 'Profit' off symbols
data['Sales'] = data['Sales'].str.replace('$', '', regex=True)
data['Sales'] = data['Sales'].str.replace(',', '', regex=True)
data['Sales'] = data['Sales'].str.replace('(', '', regex=True)
data['Sales'] = data['Sales'].str.replace(')', '', regex=True)

data['Profit'] = data['Profit'].str.replace('$', '', regex=True)
data['Profit'] = data['Profit'].str.replace(',', '', regex=True)
data['Profit'] = data['Profit'].str.replace('(', '', regex=True)
data['Profit'] = data['Profit'].str.replace(')', '', regex=True)

print(data['Sales'][:10])

print(data['Profit'][:10])
```

```python
#converting the columns into numeric and to 'int'
data['Sales'] = pd.to_numeric(data['Sales'])
data['Profit'] = pd.to_numeric(data['Profit'])

data['Sales'] = data['Sales'].astype(int)
data['Profit'] = data['Profit'].astype(int)

print(type(data['Sales'][0]))
print(type(data['Profit'][0]))
```
